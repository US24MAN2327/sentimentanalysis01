```python
import pandas as pd
text = pd.read_excel("/content/Bilawal Bhutto's tagged Tweets.xlsx")

sentiment_scores = []
response = []
for text in text['tweet']:
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    sentiment_scores.append(sentiment)
    if sentiment>0:
      response.append('POSITIVE')
    elif sentiment<0:
      response.append('NEGATIVE')
    else:
      response.append('NEUTRAL')


# Print the resulting Pandas DataFrame
print(sentiment_scores)
print(response)
```

```python
from collections import Counter
# Combine the two lists into pairs
sentiment_scores2 = [round(number,2) for number in sentiment_scores]
combined_pairs = list(zip(response, sentiment_scores2))

# Create a Counter object to count occurrences
frequency_counter = Counter(combined_pairs)

# Print the frequency table
print("Frequency Table:")
print("---------------")
print("RESPONSE\tSENTIMENT SCORES\tCOUNT")
print("-----------------------")
for pair, count in frequency_counter.items():
    print(f"{pair[0]}\t\t{pair[1]}\t\t{count}")



```

```python
import matplotlib.pyplot as plt
Explode = (0.1,0.0)
fig = plt.figure(figsize = (12,10))
df = pd.DataFrame(response, columns=['output'])
plt.pie(df['output'].value_counts(), labels = ["NEUTRAL","POSITIVE", "NEGATIVE"], autopct = "%1.1f%%", shadow = True)
plt.title('DISTRIBUTION OF SENTIMENTS', loc = 'left',fontsize = 42, color = 'green')
plt.show()
```
